# Crawling PTA

```python
import requests
from bs4 import BeautifulSoup as bs
path = "https://pta.trunojoyo.ac.id/c_search/byprod/10/"

URL = []
for i in range (1,11):
    URL.append(path+str(i))
print(URL)

arr_link_dokumen = []
for link in URL:
    resp = requests.get(link)
    soup = bs(resp.text, "lxml")
    elements = soup.find_all("a", attrs={"class":"gray button"})
    for link_dokumen in elements:
        arr_link_dokumen.append(link_dokumen.get('href'))
```

```python
import pandas as pd
import re
titles = []
writers= []
abstracts = []
PembimbingI = []
PembimbingII = []

for link_dokumen in arr_link_dokumen:
    resp = requests.get(link_dokumen)
    soup = bs(resp.text, "lxml")
    writer = soup.find("span", string=re.compile("Penulis"))
    DPI = soup.find("span", string=re.compile("Dosen Pembimbing I"))
    DPII = soup.find("span", string=re.compile("Dosen Pembimbing II"))
    abstract = soup.find("p", attrs={"align":"justify"})
    title = soup.find("a", attrs={"class","title"})
    writers.append(writer.text.replace('Penulis :',''))
    PembimbingI.append(DPI.text.replace('Dosen Pembimbing I :',''))
    PembimbingII.append(DPII.text.replace('Dosen Pembimbing II :',''))
    abstracts.append(abstract.text)
    titles.append(title.text.upper())

print(len(abstracts),len(titles),len(writers),len(PembimbingI),len(PembimbingII))
data = {'Judul':titles,'Penulis':writers, 'Dosen Pembimbing I' : PembimbingI, 'Dosen Pembimbing II' : PembimbingII, 'Abstrak':abstracts}
data
```
