# Crawling PTA

```python
import requests
from bs4 import BeautifulSoup as bs
path = "https://pta.trunojoyo.ac.id/c_search/byprod/10/"

URL = []
for i in range (1,11):
    URL.append(path+str(i))
print(URL)

arr_link_dokumen = []
for link in URL:
    resp = requests.get(link)
    soup = bs(resp.text, "lxml")
    elements = soup.find_all("a", attrs={"class":"gray button"})
    for link_dokumen in elements:
        arr_link_dokumen.append(link_dokumen.get('href'))
```

```python
import pandas as pd
import re
titles = []
writers= []
abstracts = []
PembimbingI = []
PembimbingII = []

for link_dokumen in arr_link_dokumen:
    resp = requests.get(link_dokumen)
    soup = bs(resp.text, "lxml")
    writer = soup.find("span", string=re.compile("Penulis"))
    DPI = soup.find("span", string=re.compile("Dosen Pembimbing I"))
    DPII = soup.find("span", string=re.compile("Dosen Pembimbing II"))
    abstract = soup.find("p", attrs={"align":"justify"})
    title = soup.find("a", attrs={"class","title"})
    writers.append(writer.text.replace('Penulis :',''))
    PembimbingI.append(DPI.text.replace('Dosen Pembimbing I :',''))
    PembimbingII.append(DPII.text.replace('Dosen Pembimbing II :',''))
    abstracts.append(abstract.text)
    titles.append(title.text.upper())

print(len(abstracts),len(titles),len(writers),len(PembimbingI),len(PembimbingII))
data = {'Judul':titles,'Penulis':writers, 'Dosen Pembimbing I' : PembimbingI, 'Dosen Pembimbing II' : PembimbingII, 'Abstrak':abstracts}
data
```

```python
#case folding
lower_case = show_csv['Abstrak'].str.lower()
data_lower_case = pd.concat([pd.DataFrame(lower_case, columns=["Abstrak"])], axis=1)
data_lower_case
```

```python
#cleaning
import re
clean =[]
for i in range (len(lower_case)):
  clean_mention  = re.sub("@[A-Za-z0-9_]+","", lower_case[i]) #clenasing mention
  clean_hashtag = re.sub("#[A-Za-z0-9_]+","", clean_mention) #clenasing hashtag
  clean_url = re.sub(r'http\S+', '', clean_hashtag) #cleansing url link
  clean_char = re.sub("[^a-zA-Z ]+"," ", clean_url) #cleansing character
  clean.append(clean_char)
clean_result = pd.DataFrame(clean,columns=['Hasil Cleaning'])
clean_result
```

```python
#slangwords

slang_dict = pd.read_csv("https://raw.githubusercontent.com/louisowen6/NLP_bahasa_resources/master/combined_slang_words.txt", sep=" ", header=None)

# Membuat fungsi kamus
def replace_slang_words(teks):
  kata = nltk.word_tokenize(teks.lower())
  kata_terfilter = [k for k in kata if k not in stopwords.words('indonesian')]
  for i in range(len(kata_terfilter)):
    if kata_terfilter[i] in slang_dict:
      kata_terfilter[i] = slang_dict[kata_terfilter[i]]
  return ' '.join(kata_terfilter)

# Contoh penggunaan

slang_words=[]
for i in range(len(clean)):
  slang = replace_slang_words(clean[i])
  slang_words.append(slang)

data_slang = pd.DataFrame(slang_words, columns=["Koreksi Kata"])
data_slang
```

```python
def nltk_frequency_vectorize(corpus):

    # The NLTK frequency vectorize method
    from collections import defaultdict

    def vectorize(doc):
        features = defaultdict(int)

        for token in tokenize(doc):
            features[token] += 1

        return features

    return map(vectorize, corpus)
vectnltk=nltk_frequency_vectorize(corpus)
type(vectnltk)
```
